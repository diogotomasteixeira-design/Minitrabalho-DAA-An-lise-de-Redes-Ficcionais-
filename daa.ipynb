Inicialização

```python
class Graph:
    def __init__(self):
        self.adj = {}

    def get_vertices(self):
        return list(self.adj.keys())

    def add_vertex(self, vertex):
        if vertex not in self.adj:
            self.adj[vertex] = {}

    def add_edge(self, vertex, neighbour_vertex, weight):
        self.add_vertex(vertex)
        self.add_vertex(neighbour_vertex)
        self._dj[vertex][neighbour_vertex] = weight
        self.adj[neighbour_vertex][vertex] = weight  # não orientado

    def get_neighbors(self, v):
        return list(self.adj[v].items())

    def num_vertices(self):
        return len(self.adj)

    def num_edges(self):
        return sum(len(n) for n in self.adj.values())
```

BFS

```python
from collections import deque


def bfs(graph, source):
    visited = {}
    dist = {}
    parent = {}

    for v in graph:
        visited[v] = False
        dist[v] = float('inf')
        parent[v] = None

    queue = deque()

    visited[source] = True
    dist[source] = 0
    queue.append(source)

    while queue:
        v = queue.popleft()

        for w in graph[v]:
            if not visited[w]:
                visited[w] = True
                dist[w] = dist[v] + 1
                parent[w] = v
                queue.append(w)

    return visited, dist, parent

#############################################

graph = {
    "Eddard-Stark": ["Jon-Snow", "Robert-Baratheon", "Catelyn-Stark"],
    "Jon-Snow": ["Eddard-Stark", "Samwell-Tarly"],
    "Robert-Baratheon": ["Eddard-Stark", "Cersei-Lannister"],
    "Catelyn-Stark": ["Eddard-Stark", "Arya-Stark", "Sansa-Stark"],
    "Samwell-Tarly": ["Jon-Snow"],
    "Cersei-Lannister": ["Robert-Baratheon", "Jaime-Lannister"],
    "Jaime-Lannister": ["Cersei-Lannister"],
    "Arya-Stark": ["Catelyn-Stark"],
    "Sansa-Stark": ["Catelyn-Stark"]
}


visited, dist, parent = bfs(graph, "Jon-Snow")

print("=== VISITED ===")
for v in visited:
    print(v, "->", visited[v])

print("\n=== DISTÂNCIAS ===")
for v in dist:
    print(v, "->", dist[v])

print("\n=== PARENTS / CAMINHO ===")
for v in parent:
    print(v, "->", parent[v])
```

Conectividade

```python
def num_components(graph):
    checked = set()
    count = 0
    total_vertices = graph.get_vertices()

    for vertex in total_vertices:
        if vertex not in checked:
            components = graph.bfs(vertex)
            checked.update(components)
            count += 1

    return count
```

```python
def largest_component(graph):
    checked = set()
    component = []
    total_vertices = graph.get_vertices()

    for vertex in total_vertices:
        if vertex not in checked:
            components = graph.bfs(vertex)
            checked.update(components)
            component.append(components)

    largest_component = max(component, key=len)
    
    new_graph = Graph()
    for vertex in largest_component:
        new_graph.add_vertex(vertex)
    for vertex in largest_component:
        for neighbour_vertex, weight in graph.get_neighbours(vertex):
            if neighbour_vertex in largest_component:
                new_graph.add_edge(vertex, neighbour_vertex, weight)

    return new_graph
```

```python
def degree_centrality(graph):
    n = graph.num_vertices()
    DC = {}

    for vertex in range (graph.get_vertices()):
        grau = len(graph.adj)
        DC[v] = grau / (n-1)

    return DC


def closenness_centrality(graph):
    n = graph.num_vertices()
    CC = {}

    for vertex in graph.get_vertices():
        distances = graph.bfs_distances(vertex)

        rv = len(distances)
        soma = sum(distances.values())

        if rv == 0 or soma == 0:
            CC[v] = 0.0
        else:
            CC[v] = (rv ** 2) / ((n - 1) * soma)

    return CC
```
